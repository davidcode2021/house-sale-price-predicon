# checking for outliers and replacing them with mean value of the features

```python
numerical_columns = [column for column in df.columns if df[column].dtype==int and len(df[column].unique()<11)]
numerical_columns=numerical_columns[:-1].copy()
```

```python
def outlier_detection(column, df):
    outlier=[]
    for x in  df[column]:
        z=np.divide(x-np.mean(df[column]),np.std(df[column]))
        if z > 3:
            outlier.append(x)
            df[column].replace(x,math.ceil(np.mean(df[column])),inplace=True)
#     print(column, outlier)     
    

for column in numerical_columns:
    outlier_detection(column, df)
```

# converting categorical features to dummy variables

```python
categorical_columns = [column for column in df.columns if df[column].dtype=="object"]
# categorical_columns
```

```python
df1=pd.get_dummies(data=df,columns=categorical_columns,dummy_na=True)
df_new=pd.concat([df1,df["SalePrice"]], axis=1)
```

# preparing the data for modeling

```python
# spliting the data into dependent and indepent variable
X=df_new.iloc[:,:-1]
y=df_new.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
# scaling the data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(300)
X_pca=pca.fit_transform(X)
```

```python
# pca.explained_variance_ratio_
X_pca.shape
```

# modeling

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.svm import SVR
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
svm_classifier=SVR(kernel="rbf", )
random_regressor=RandomForestRegressor(n_estimators=200)
logistic_regression=LogisticRegression()
```

```python
num_kfold=5
kf=KFold(n_splits=num_kfold,shuffle=True,random_state=32)
```

```python
cross_val_result=cross_val_score(random_regressor, X_pca,y ,cv=kf,scoring='r2')

print("Accuracy:",cross_val_result)
print("Mean Accuracy:",np.mean(cross_val_result))                                       
```
